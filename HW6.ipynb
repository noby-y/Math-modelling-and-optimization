#### $f(x) = d_1 * x^3 + d_2 * x^2 + d_3 * x + d_4 \rightarrow min$
#### на інтервалі $х \in [0;10]$ з точністю $\epsilon = 0,01$:
1) методом половинного ділення інтервалу;

Value table: 
|$d_1$|$d_2$|$d_3$|$d_4$| 
| :-: | :-: | :-: | :-: |
| $1$ | $-3$|$-30$|$140$|

$f(x) = x^3 - 3 * x^2 - 30 * x + 140 \rightarrow min$

## 1) Метод половинного ділення інтервалу

```python
def f(x):
    return x**3 - 3*x**2 - 30*x + 140

def bisection_method(a, b, tol=1e-6):
    iterations = 0
    while abs(b - a) > tol:
        iterations += 1
        mid = (a + b) / 2
        if f(mid - tol) < f(mid + tol):
            b = mid
        else:
            a = mid
        print(f"Iteration {iterations:>2}: a = {a:<16.4f} b = {b:<16.4f} x = {(a + b) / 2:<16.4f} |a-b| = {(abs(a - b)):<16.4f}")

    return (a + b) / 2

# Interval [a, b]
a = 0 
b = 10

# Tolerance
e = 0.01 

min_point = bisection_method(a, b, e)
min_value = f(min_point)

print("-"*92)
print("ANSWER")
print("\nMinimum point:", min_point)
print("Minimum value:", min_value)

x_values = np.linspace(a, b, 100)
y_values = f(x_values)

# Plot the function
# plt.figure(figsize=(8, 6))
# plt.plot(x_values, y_values, label='Function: $f(x) = x^3 - 3 * x^2 - 30 * x + 140 \\rightarrow min$')
# plt.scatter(min_point, min_value, color='red', label='Minimum Point')
# plt.xlabel('x')
# plt.ylabel('f(x)')
# plt.title('Function Plot with Minimum Point')
# plt.legend()
# plt.grid(True)
# plt.show()
```

## 2) Метод дихотомії або метод золотого перерізу або метод Фібоначчі.

```python
def golden_section_method(a, b, tol=1e-6):
    gr = (np.sqrt(5) + 1) / 2  # Golden ratio
    
    x1 = b - (b - a) / gr
    x2 = a + (b - a) / gr
    
    iterations = 0
    while abs(b - a) > tol:
        if f(x1) < f(x2):
            b = x2
        else:
            a = x1
        
        x1 = b - (b - a) / gr
        x2 = a + (b - a) / gr
        
        iterations += 1
        print(f"Iteration {iterations:>2}: a = {a:<16.4f} b = {b:<16.4f} x1 = {x1:<16.4f} x2 = {x2:<16.4f}")
    
    return (b + a) / 2


# Perform optimization using the golden section method
min_point = golden_section_method(a, b)
min_value = f(min_point)

print("-"*92)
print("ANSWER")
print("\nMinimum point:", min_point)
print("Minimum value:", min_value)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Function: $f(x) = x^3 - 3 * x^2 - 30 * x + 140 \\rightarrow min$')
plt.scatter(min_point, min_value, color='red', label='Minimum Point')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function Plot with Minimum Point')
plt.legend()
plt.grid(True)
plt.show()
```
