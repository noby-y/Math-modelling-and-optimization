## 1.
Випадкова функція $X(t)=U·sin(t)$, де $U$ – випадкова величина, рівномірно
розподілена на відрізку $[-1;1]$. Вказати кількість можливих реалізацій

## 5.
Випадковий процес заданий функцією $X(t) = t + U * cos(t)$, де $U$ – випадкова
величина з математичним сподіванням, рівним 2, та дисперсією, рівною 1. Знайти
кореляційну функцію та нормовану кореляційну функцію випадкового процесу
$X(t)$.

$M(U) = 2$  
$D(U) = 1$

#### Кореляційна функція:

$K_x(t_1, t_2) = M[(X(t_1) - m_x(t_1)) * (X(t_2) - m_x(t_2))]$  

$m_x(t) = M[t + U * cos(t)] = t + cos(t) * M[U] = t + 2cos(t)$  
$m_x(t_1) = t_1 + 2cos(t_1)$  
$m_x(t_2) = t_2 + 2cos(t_2)$  

$X(t_1) = t_1 + U * cos(t_1)$  
$X(t_2) = t_2 + U * cos(t_2)$  

$K_x(t_1, t_2) = M[(t_1 + U * cos(t_1) - t_1 - 2cos(t_1)) * (t_2 + U * cos(t_2) - t_1 - 2cos(t_1))] = M[cos(t_1) * cos(t_2) * (U - 2)^2] = cos(t_1) * cos(t_2) * D[U] = cos(t_1) * cos(t_2)$  

#### Нормована кореляційна функція:

$r_x(t_1, t_2) = \frac{K_x(t_1, t_2)}{\sqrt{K_x(t_1, t_1)} * \sqrt{K_x(t_2, t_2)}} $  

$\sqrt{K_x(t_1, t_1)} = \sqrt{cos(t_1) * cos(t_1)} = cos(t_1)$  
$\sqrt{K_x(t_2, t_2)} = \sqrt{cos(t_2) * cos(t_2)} = cos(t_2)$  

$r_x(t_1, t_2) = \frac{cos(t_1) * cos(t_2)}{cos(t_1) * cos(t_2)} = 1$

1) 3 виклики;
2) менш як 3 виклики;
3) не менш як 3 виклики.

Потік викликів вважається найпростішим.

$\lambda = 2$  
$t = 4$

$P_t(m) = \frac{(\lambda t)^m}{m!} * e^{-\lambda t} = \frac{(8)^m}{m!} * e^{-8}$

#### 1) $m_1 = 3$

$P_t(m_1) = \frac{(2 * 4)^3}{3!} * e^{-2 * 4} = \frac{512}{6} * e^{-8} \approx 85.(3) * 0.0003354626 \approx 0.0286261419$

#### 2) $m_2 = [0, 2]$

$P_t(0) = \frac{(8)^0}{0!} * e^{-8} = 1 * e^{-8} \approx 0.00033546$  
$P_t(1) = \frac{(8)^1}{1!} * e^{-8} = 8 * 0.00033546 \approx 0.0026837008$  
$P_t(2) = \frac{(8)^2}{2!} * e^{-8} = 32 * 0.00033546 \approx 0.0107348032$

$P_t(m_2) = P_t(0) + P_t(1) + P_t(2) = 0.00033546 + 0.0026837008 + 0.0107348032 \approx 0.013753964$ 

#### 3) $m_3 \geq 3$

$P_t(m_3) = 1 - P_t(m_2) = 1 - 0.0286261419 = 0.9713738581$
