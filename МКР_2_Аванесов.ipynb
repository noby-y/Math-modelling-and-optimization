```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 2*(x**3) - 2*(x**2) - 50*(x) + 130

def midpoint_method(a, b, tol=1e-6):
    iterations = 0
    while abs(b - a) > tol:
        iterations += 1
        mid = (a + b) / 2
        if f(mid - tol) < f(mid + tol):
            b = mid
        else:
            a = mid
        print(f"Iteration {iterations:<2}: a = {a:<16}, b = {b:<16}, x = {(a + b) / 2:<16}")
    return (a + b) / 2

# interval [a, b]
a = 0
b = 10
# tolerance e
e = 0.01

min_point = midpoint_method(a, b, tol=e)
min_value = f(min_point)

print("Minimum point:", min_point)
print("Minimum value:", min_value)

# Generate points for plotting the function
x_values = np.linspace(a, b, 100)
y_values = f(x_values)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Function: $f(x) = x^3 - 3x^2 + -30x + 140$')
plt.scatter(min_point, min_value, color='red', label='Minimum Point')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function Plot with Minimum Point')
plt.legend()
plt.grid(True)
plt.show()
```
