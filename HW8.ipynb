#### Задача 1.
Знайти оптимальні значення змінних $x_1$* та $x_2$\*, такі, що приводять цільову функцію до екстремуму при заданих обмеженнях. Обчислити значення цільової функції $F(x_1*, x_2*)$. Задачу розв’язати графічним методом.

$
F = 4x_1 + 6x_2 \rightarrow max \\
$  
$
\begin{cases}
2x_1 + 4x_2 <= 16 \\
x_1 + x_2 <= 6 \\
x_1 + 4x_2 <= 14 
\end{cases} \\
$  
$
x_1 >= 0,\  x_2 >= 0
$

Відповідь:
1) оптимальні значення параметрів $х_1$* = 4; $х_2$* = 2;
2) найбільше значення цільової функції $F(х_1*, х_2*)$ = 28.0.


```python
# Define the inequalities
x = np.linspace(0, 10, 400)
bound_1 = (16 - 2*x) / 4
bound_2 = 6 - x
bound_3 = (14 - x) / 4
F = 4 * x / 6

plt.plot(x, bound_1, label=r'$bound_1 →\ 2x_1 + 4x_2 \leq 16$')
plt.plot(x, bound_2, label=r'$bound_2 →\ x_1 + x_2 \leq 6$')
plt.plot(x, bound_3, label=r'$bound_3 →\ x_1 + 4x_2 \leq 14$')

plt.fill_between(x, np.minimum(np.minimum(bound_1, bound_2), bound_3), 0, where=(x >= 0) & (bound_1 >= 0) & (bound_2 >= 0) & (bound_3 >= 0), color='grey', alpha=0.3)

plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.title('Region of solutions')
plt.legend()

plt.grid(True)
plt.xlim(0, 9)
plt.ylim(0, 9)
plt.show()
```

#### Точки перетину:

\end{cases} \\  
2x_1 - 2x_1 + 4x_2 - 2x_2 = 16 - 6 * 2 \\

```python
corner_points = np.array([[0, 3.5], [2, 3], [4, 2], [6, 0]])

def F(x1, x2):
    return 4*x1 + 6*x2

max_value = float('-inf')
max_point = None

for point in corner_points:
    value = F(point[0], point[1])
    if value > max_value:
        max_value = value
        max_point = point

print(f"Max value: {max_value} in point {[coord for coord in max_point]}")
```
